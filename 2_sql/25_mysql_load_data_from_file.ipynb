## NYSE Daily Stock Prices 
The dataset contains daily OHLC (Open, High, Low, Close) prices along with volume and adjusted close values.

## Dataset Structure 

Each row in the CSV file contains:

1. Exchange (e.g., NYSE)
2. Stock Symbol (e.g., QTM)
3. Trade Date
4. Open Price
5. High Price
6. Low Price
7. Close Price
8. Volume
9. Adjusted Close Price

### Create Database

````python
CREATE DATABASE IF NOT EXISTS nyse;
````

````python
USE nyse;
````

### Create Table (DDL)

````python
CREATE TABLE nyse_daily_prices (
    exchange        VARCHAR(10),
    symbol          VARCHAR(10),
    trade_date      DATE,
    open_price      DECIMAL(10,2),
    high_price      DECIMAL(10,2),
    low_price       DECIMAL(10,2),
    close_price     DECIMAL(10,2),
    volume          BIGINT,
    adj_close_price DECIMAL(10,2)
);
````
