```python
USE demo1;
```


## Create DEPT Table


```python

CREATE TABLE dept (
    deptno INT,
    dname  VARCHAR(14),
    loc    VARCHAR(13)
);
```


## Create EMP Table


```python

CREATE TABLE emp (
    empno    INT,
    ename    VARCHAR(10),
    job      VARCHAR(9),
    mgr      INT,
    hiredate DATE ,
    sal      DECIMAL(7,2) ,
    comm     DECIMAL(7,2),
    deptno   INT,
);
```
