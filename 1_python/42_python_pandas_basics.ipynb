## ðŸ§± Importing Pandas


````
import pandas as pd
````

## ðŸ§® Creating a DataFrame


````
data = {
    'Name': ['Arjun', 'Ravi', 'Priya'],
    'Age': [20, 30, 21],
    'City': ['Bengaluru', 'Mumbai', 'Mysuru']
}

df = pd.DataFrame(data)
print(df)
````

## ðŸŽ¯ Accessing Columns


````
print(df['Name'])
print(df[['Name', 'City']])
````

## ðŸ” Filtering Rows


````
# Filter rows where age > 20
print(df[df['Age'] > 20])
````
