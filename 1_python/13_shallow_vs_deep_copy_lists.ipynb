# Shallow Copy vs Deep Copy in Python

This notebook explains **shallow copy** and **deep copy** using simple, observable steps.

**Analogy:**
- Outer list = outer box
- Nested list = box inside the box

## Flat List

```python
l1 = [1, 2, 3]
l2 = l1.copy()

print('Before change:')
print('l1:', l1)
print('l2:', l2)

l1.append(4)

print('\nAfter l1.append(4):')
print('l1:', l1)
print('l2:', l2)
```

## Nested List + Shallow Copy

```python
l1 = [1, 2, [3, 4]]
l2 = l1.copy()

print('Before change:')
print('l1:', l1)
print('l2:', l2)

l1[2].append(99)

print('\nAfter l1[2].append(99):')
print('l1:', l1)
print('l2:', l2)

l1.append(4)

print('\nAfter l1.append(4):')
print('l1:', l1)
print('l2:', l2)
```

## Nested List + Deep Copy

```python
import copy

l1 = [1, 2, [3, 4]]
l3 = copy.deepcopy(l1)

print('Before change:')
print('l1:', l1)
print('l3:', l3)

l1[2].append(99)

print('\nAfter l1[2].append(99):')
print('l1:', l1)
print('l3:', l3)

l1.append(4)

print('\nAfter l1.append(4):')
print('l1:', l1)
print('l3:', l3)
```

## Key Takeaway

- Shallow copy: outer object is new, inner mutable objects are shared
- Deep copy: both outer and inner objects are independent
